```python
df['type'].value_counts().plot(kind = 'bar',xlabel="Type", ylabel ='count',title = 'Types of Restaurent');
```

```python
top_cuisines = df['cuisines'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cuisines.index,y=top_cuisines.values)
plt.xticks(rotation=45,ha='right')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.title('Top 10 Cuisines')
plt.show()
```

```python
online_order = df['online_order'].value_counts()
plt.figure(figsize=(7,7))
sns.barplot(x=online_order.index, y = online_order.values)
plt.xlabel('online_order')
plt.ylabel('Count')
plt.title('online_order Count')
plt.show()
```

```python
book_table = df['book_table'].value_counts()
plt.figure(figsize = (6,6))
sns.barplot(x=book_table.index, y = book_table.values)
plt.xlabel('book_table')
plt.ylabel('Count')
plt.title('book Table Count')
plt.show()
```
